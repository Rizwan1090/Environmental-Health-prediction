#Environmental Health Monitoring System

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
# Load the datasets
pune_df = pd.read_csv('Punedata.csv')
health_df = pd.read_csv('Healthdata.csv')


```

```python
pune_df.head(5)
```

```python
# Fill missing values with the median for each column in MH014.csv dataset
pune_df_filled = pune_df.fillna(pune_df.median(numeric_only=True))
```

```python
# Drop columns with excessive missing values
columns_to_drop = ["Toluene (ug/m3)", "Xylene (ug/m3)", "RF (mm)"]
pune_df_cleaned = pune_df_filled.drop(columns=columns_to_drop)

```

```python
# Extract date features
pune_df_cleaned['From Date'] = pd.to_datetime(pune_df_cleaned['From Date'])
pune_df_cleaned['To Date'] = pd.to_datetime(pune_df_cleaned['To Date'])
pune_df_cleaned['Hour'] = pune_df_cleaned['From Date'].dt.hour
pune_df_cleaned['Day'] = pune_df_cleaned['From Date'].dt.day
pune_df_cleaned['Month'] = pune_df_cleaned['From Date'].dt.month
pune_df_cleaned['Year'] = pune_df_cleaned['From Date'].dt.year

pune_df_cleaned.head(5)
```

```python
pune_df_cleaned.head(5)
```

```python

# Merging environmental data with health data
combined_df = pd.concat([pune_df_cleaned, health_df], axis=1)
combined_df.head(5)
```

```python
combined_df = combined_df.fillna(combined_df.median(numeric_only=True))
```

```python
# Calculate correlation matrix
correlation_matrix = combined_df.corr()

correlation_matrix
```

```python
# Plot correlation matrix
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix Between Environmental Factors and Health Symptoms')
plt.show()
```

```python
# Define features and target variable
features = combined_df.drop(columns=['Tiredness', 'Dry-Cough', 'Difficulty-in-Breathing', 'Sore-Throat', 'None_Sympton', 'Pains', 'Nasal-Congestion', 'Runny-Nose', 'None_Experiencing'])
features = features.apply(pd.to_numeric, errors='coerce')
features = features.fillna(features.median(numeric_only=True))
targets = combined_df[['Tiredness', 'Dry-Cough', 'Difficulty-in-Breathing', 'Sore-Throat', 'None_Sympton', 'Pains', 'Nasal-Congestion', 'Runny-Nose', 'None_Experiencing']]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=42)

```

```python
# Initialize and train the model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
features = combined_df.drop(columns=['Tiredness', 'Dry-Cough', 'Difficulty-in-Breathing', 'Sore-Throat', 'None_Sympton', 'Pains', 'Nasal-Congestion', 'Runny-Nose', 'None_Experiencing'])
features = features.apply(pd.to_numeric, errors='coerce')
features = features.fillna(features.median(numeric_only=True))

# Define target variable
targets = combined_df[['Tiredness', 'Dry-Cough', 'Difficulty-in-Breathing', 'Sore-Throat', 'None_Sympton', 'Pains', 'Nasal-Congestion', 'Runny-Nose', 'None_Experiencing']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, random_state=42)

# Initialize and train the model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
```

```python
X_train.columns
```

```python
y_pred
```

```python
# Evaluate the model
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred, multioutput='variance_weighted')

print(f'RMSE: {rmse}')
print(f'R²: {r2}')
```

```python
# Plot actual vs predicted values for one of the health symptoms
plt.figure(figsize=(10, 6))
plt.scatter(y_test['Tiredness'], y_pred[:, 0], alpha=0.3)
plt.xlabel('Actual Tiredness')
plt.ylabel('Predicted Tiredness')
plt.title('Actual vs Predicted Tiredness')
plt.show()
```

```python
# Feature importance plot
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]
feature_names = features.columns
```

```python
plt.figure(figsize=(12, 6))
plt.title("Feature Importances")
plt.bar(range(X_train.shape[1]), importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), feature_names[indices], rotation=90)
plt.show()
```
